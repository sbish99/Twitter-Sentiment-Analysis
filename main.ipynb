```python
import pandas as pd
import numpy as np
from scipy.special import softmax
import re
from langdetect import detect
from transformers import AutoTokenizer, AutoModelForTokenClassification, AutoConfig, AutoModelForSequenceClassification
```

```python
def check_language(sentence):
    try:
        language = detect(sentence)
        if language == 'en':
            return sentence
        else:
            return "not in English"
    except:
        return "not in English"

def clean_tweet_text(tweet):
    # Replace hyperlinks with "http"
    tweet = re.sub(r'http\S+', 'http', tweet)
    # Replace user mentions with "@user"
    tweet = re.sub(r'@\w+', '@user', tweet)
    return tweet

data = pd.read_csv('chatgpt1.csv')
data.head()
# data.isnull().sum()
```

```python
sub_data = data.head(10)
sub_data['Text'] = sub_data['Text'].apply(clean_tweet_text)
sub_data['Text'] = sub_data['Text'].apply(check_language)
sub_data.head(10)
```

```python
Model_link = 'cardiffnlp/twitter-roberta-base-sentiment-latest'

tokenizer = AutoTokenizer.from_pretrained(Model_link)
model = AutoModelForSequenceClassification.from_pretrained(Model_link)
config = AutoConfig.from_pretrained(Model_link)

txt = sub_data.iloc[1,2]
# txt = "This is so good!"
print(txt)

labels = ['Negative', 'Neutral', 'Positive']

encoded_tweet = tokenizer(txt, return_tensors='pt')
print(encoded_tweet)

output = model(encoded_tweet['input_ids'], encoded_tweet['attention_mask'])
# output = model(**encoded_tweet)
print(output)

scores = output[0][0].detach().numpy()
print(scores)
scores = softmax(scores)
print(scores)

# ranking = np.argsort(scores)
# ranking = ranking[::-1]
# for i in range(scores.shape[0]):
#     l = config.id2label[ranking[i]]
#     s = scores[ranking[i]]
#     print(f"{i+1}) {l} {np.round(float(s), 4)}")
for i in range(len(scores)):
    
    l = labels[i]
    s = scores[i]
    print(l,s)
```
